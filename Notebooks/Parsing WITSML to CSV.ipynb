```python
import pandas as pd
import xml.etree.ElementTree as ET

def parse_witsml_drilling(xml_file):
    # Load and parse the XML
    tree = ET.parse(xml_file)
    root = tree.getroot()
    
    # Find namespaces
    data_list = []
    
    # Locate the mnemonic header, usually found in <mnemonicList>
    ns = {'witsml': 'http://www.witsml.org/schemas/1series'} # Common WITSML namespace
    
    # Find the data rows (usually in <data> tags)
    for log_data in root.findall('.//witsml:logData', ns):
        mnemonics = log_data.find('witsml:mnemonicList', ns).text.split(',')
        units = log_data.find('witsml:unitList', ns).text.split(',')
        
        for data_row in log_data.findall('witsml:data', ns):
            values = data_row.text.split(',')
            data_list.append(dict(zip(mnemonics, values)))

    # Create DataFrame
    df = pd.DataFrame(data_list)
    
    # Convert numeric columns from strings
    for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='ignore')
        
    return df, mnemonics

# Run the parser on WITSML file
# Example: df_drilling, columns = parse_witsml_drilling('15_9-F-12 (W-353084).xml')

# Export to CSV
df_drilling.to_csv('file_name.csv', index=False)
```
